```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car-sales.csv")
```

```python
df
```

```python
colours = df['Colour'].unique()
colours
```

```python
df_colours = pd.DataFrame()
def One_hot_encoding(series,name):
    return (series == name).astype(int)

for i in colours:
    df_colours[i] = One_hot_encoding(df['Colour'],i)
```
