```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Reading and displaying Images

```python
img = cv.imread('Photos/cat.jpg')
cv.imshow("Billu",img)
cv.waitKey(0)
# reading images
```

```python
def inline_img(img,axis = "off"):
    img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    if (axis!="off"):
        pass
    else:
        plt.axis("off")
    plt.show()
inline_img(img)
# Function to print images inline in notebook
```

## Resizing and cropping

```python
resized = cv.resize(img,(224,224))
inline_img(resized)
```

```python
inline_img(img,"s")
```

```python
#cropping
cropped = img[50:,300:]
inline_img(cropped)
```

```python
#cropping to a square
height, width = cropped.shape[:2]
size = min (height, width)
x = (width - size) // 2
y = (height - size) // 2
sqcropped = cropped[y: y+size, x:x+size]
inline_img(sqcropped)
```

## Converting between colorspaces

```python
# Convert images to greyscale and other colorcpaces
rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
inline_img(gray)
inline_img(rgb)
```

```python
# THRESHOLDING
ret, thresh = cv.threshold(gray,40,255,cv.THRESH_BINARY)
inline_img(thresh)
```

```python
thresh = cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,31,5)
inline_img(thresh)
```

## EDGE DETECTION

```python
img_edge = cv.Canny(blurred,50,50) #parametes through trial and error
inline_img(img_edge)
```

# PREPROCESSING AND EDGE DETECTION OF A CAR

```python
img = cv.imread("Photos/car.png")
inline_img(img)
```

```python
img.shape
```

```python
resized = cv.resize(img,(398,224)) #values based on the shape of the image
inline_img(resized) #downscaling the image for faster calculation
```

```python
grey = cv.cvtColor(resized,cv.COLOR_BGR2GRAY)
inline_img(grey)
```

```python
blurred = cv.GaussianBlur(grey,(3,3),0)
inline_img(blurred)
```

```python
rows, colmns = blurred.shape

x_kernel = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

y_kernel = np.array([[-1, -2, -1],
                    [0,  0,  0],
                    [1,  2,  1]])

edge = np.zeros((rows, colmns))

for i in range(1, rows-1):
    for j in range(1, colmns-1):
        region = blurred[i-1:i+2, j-1:j+2]
        gx = np.sum(np.multiply(x_kernel, region))
        gy = np.sum(np.multiply(y_kernel, region))
        edge[i, j] = int(np.sqrt(gx**2 + gy**2))

plt.title('Edges')
plt.imshow(edge, cmap='gray', vmin=0, vmax=255)
plt.axis("off")
plt.show()

```
