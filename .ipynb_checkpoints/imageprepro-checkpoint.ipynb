```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

## Reading and displaying Images

```python
img = cv.imread('Photos/cat.jpg')
cv.imshow("Billu",img)
cv.waitKey(0)
# reading images
```

```python
def inline_img(img,axis = "off"):
    img_rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    if (axis!="off"):
        pass
    else:
        plt.axis("off")
    plt.show()
inline_img(img)
# Function to print images inline in notebook
```

## Resizing and cropping

```python
resized = cv.resize(img,(224,224))
inline_img(resized)
```

```python
inline_img(img,"s")
```

```python
#cropping
cropped = img[50:,300:]
inline_img(cropped)
```

```python
#cropping to a square
height, width = cropped.shape[:2]
size = min (height, width)
x = (width - size) // 2
y = (height - size) // 2
sqcropped = cropped[y: y+size, x:x+size]
inline_img(sqcropped)
```

## Converting between colorspaces

```python
# Convert images to greyscale and other colorcpaces
rgb = cv.cvtColor(img,cv.COLOR_BGR2RGB)
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
inline_img(gray)
inline_img(rgb)
```

```python
# THRESHOLDING
ret, thresh = cv.threshold(gray,40,255,cv.THRESH_BINARY)
inline_img(thresh)
```

```python
thresh = cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY,31,5)
inline_img(thresh)
```

## Blurring

#### OpenCV functions to blur an image :<br>
> blur()<br>
> GaussianBlur()<br>
> medianBlur()<br>
> bilateralFilter()<br>

```python
blurred = cv.blur(gray,(7,7))
inline_img(blurred)
```

```python
med_blurred = cv.medianBlur(gray,7)
inline_img(med_blurred)
```

## EDGE DETECTION

```python
img_edge = cv.Canny(blurred,50,50) #parametes through trial and error
inline_img(img_edge)
```

# PREPROCESSING AND EDGE DETECTION OF A CAR

```python
img = cv.imread("Photos/car.png")
inline_img(img)
```

```python
grey = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
inline_img(grey)
```

```python
blurred = cv.GaussianBlur(grey,(11,11),0)
inline_img(blurred)
```

```python
kenny = cv.Canny(blurred,50,150)
inline_img(kenny)
```

```python
kenny_d = cv.dilate(kenny,np.ones((5,5),dtype = np.int8))
inline_img(kenny_d)
```
