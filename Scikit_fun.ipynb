# LINEAR REGRESSION

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Databases/car-sales-extended.csv")
df.head()
```

#### Relation between odometer, doors and the price of the car

```python
lis = ["Doors","Odometer (KM)"]
x = df[lis]
y=  df["Price"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
predictions = model.predict(x_test)
test_residuals = y_test-predictions
```

#### Residual plot

```python
sns.scatterplot(x=y_test,y=test_residuals)
plt.axhline(y=0,color='red')
```

#### Distribution plot

```python
sns.displot(test_residuals,bins=25,kde=True)
```

# LOGISTIC REGRESSION

```python
df = pd.read_csv("Databases/heart-disease.csv")
df.head()
```

```python
x = df.drop("target",axis=1)
```

```python
y=df["target"]
```

```python
x_train, x_test, y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_x_train,y_train)
```

```python
y_pred = log_model.predict(scaled_x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

### Accuracy of the model = 88.15%
